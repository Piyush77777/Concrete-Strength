```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
train_rmse_error=[]
test_rmse_error=[]
for d in range(1,11):
    Poly_converter= PolynomialFeatures(degree=d,include_bias=False)
    Poly_features= Poly_converter.fit_transform(X)
    X_train,X_test,y_train,y_test= train_test_split(Poly_features,y,test_size=0.3, random_state=101)
    
    lnmodel=LinearRegression()
    lnmodel.fit(X_train,y_train)
    
    train_pred=lnmodel.predict(X_train)
    test_pred=lnmodel.predict(X_test)
    
    train_rmse= np.sqrt(mean_squared_error(y_train,train_pred))
    test_rmse= np.sqrt(mean_squared_error(y_test,test_pred))
    
    train_rmse_error.append(train_rmse)
    test_rmse_error.append(test_rmse)
    
    
```

```python
train_rmse_error
```

```python
test_rmse_error
```

```python
plt.plot(range(1,11),train_rmse_error,label='Train RMSE')
plt.plot(range(1,11),test_rmse_error,label='Test RMSE')
plt.ylabel("RMSE")
plt.xlabel("Degree Of Polynomial")
plt.legend()
```

```python
plt.plot(range(1,5),train_rmse_error[0:4],label='Train RMSE')
plt.plot(range(1,5),test_rmse_error[0:4],label='Test RMSE')
plt.ylabel("RMSE")
plt.xlabel("Degree Of Polynomial")
plt.legend()
```
